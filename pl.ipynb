Feature engineering for baseline model

```python
data["venue_bin"]=data["venue"].astype("category").cat.codes
```

```python
data[["venue", "venue_bin",]]
## venue is an important feature/predictor (home ground advantage), however, given it is of the object dtype
## it can't be used by ML models so we create the venue_bin feature to numerically identify each venue.
```

```python
data["opp_bin"]=data["opponent"].astype("category").cat.codes
data[["opponent", "opp_bin"]]
## opponent is an important feature/predictor, however, given it is of the object dtype
## it can't be used by ML models so we create the opp_bin feature to numerically identify each team
```

```python
data["hour_of_day"]=data["time"].str.replace(":.+", "", regex=True).astype("int")
## some teams may perform better when they play during a particualr hour of day
```

```python
data["day_bin"]=data["date"].dt.day_of_week
## similar idea as before, some teams may perform better, if they play on a particular day of the week
```

```python
data["target"]=(data["result"] == "W")
data["target"] = data["target"].astype("int")
data[["result", "target"]]
## Here we are setting our target column which is what our model will be predicting. 
## The model predicts wetcher a team will win or not, hence both Draw and Loss will be represented as zero
```

Further feature engineering

```python
team_matches=data.groupby("team")
Arsenal_group=team_matches.get_group("Arsenal")
# group the records in the data frame by teams 
# so that we can calculate the rolling averages for different features.
# Then use these additional features for improved model performance
```

```python
def rolling_avg(group, col, new_cols):
    group=group.sort_values("date")
    rolling_avgerages=group[col].rolling(3, closed='left').mean()
    group[new_cols]=rolling_avgerages
    group=group.dropna(subset=new_cols)
    return group
```

```python
columns = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_columns= ["rol_gf", "rol_ga", "rol_sh", "rol_sot", "rol_dist", "rol_fk", "rol_pk", "rol_pkatt"]
# new_columns=[f"rol_{c}" for c in columns]
```

```python
data_new=data.groupby("team").apply(lambda x: rolling_avg(x, columns, new_columns))
```

```python
data_new
```

```python
data_new=data_new.droplevel("team")
```

```python
data_new.index=range(data_new.shape[0])
```

Retraining machine learning model

```python
def predtion_testing(data, predictors):
    train=data[data["date"]<'2022-01-01']
    test=data[data["date"]>'2022-01-01']
    model.fit(train[predictors], train["target"])
    predictions=model.predict(test[predictors])
    resultant_frame2=pd.DataFrame(dict(actual=test["target"], predicted=predictions), index=test.index)
    precision=precision_score(test["target"], predictions)
    acc=accuracy_score(test["target"], predictions)
    return resultant_frame2, acc, precision
```

```python
resultant_frame2, accuracy_result, precs=predtion_testing(data_new,features+new_columns)
```

```python
accuracy_result
```

```python
precs
# Incorporating rolling averages from the previous three matches significantly improved model performance, 
# increasing precision from 47% to 62.5%. 
# Rolling features capture recent form and short-term momentum, 
# allowing the model to reflect current team performance more accurately than static season-level averages.
```
